```python
# import necessary libraries - data manipulation and visualization
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load a sample dataset - Restaurant Tips Dataset
csv_path = "./Dataset/Restaurant-Tips.csv"
df = pd.read_csv(csv_path)

print("\n‚úÖ Dataset loaded successfully!")
print(df.head())
```

### üß© 1Ô∏è‚É£ UNIVARIATE PLOTS
Univariate plots display information about a SINGLE variable.      
They help us understand data distribution, spread, and outliers.

üìà Common Univariate Plots:
- Histogram
- Density Plot
- Box Plot

```python
# Histogram ‚Äî shows frequency distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['total_bill'], bins=15, kde=False, color='skyblue')
plt.title('üå∏ Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()

print("\n‚úÖ Histogram displayed successfully!")
```

```python
# Density Plot ‚Äî smooth version of histogram - Dataset restaurant tips
plt.figure(figsize=(8, 5))
sns.kdeplot(df['total_bill'], shade=True, color='orange')
plt.title('üå∏ Density Plot of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.show()
print("\n‚úÖ Density plot displayed successfully!")
```

```python
# Box Plot ‚Äî shows median, quartiles, and outliers
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['total_bill'], color='skyblue')
plt.title('üå∏ Box Plot of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()

print("\n‚úÖ Box plot displayed successfully!")
```

### üß© 2Ô∏è‚É£ MULTIVARIATE PLOTS
Multivariate plots show relationships between TWO or MORE variables.

üìä Common Multivariate Plots:
- Scatter Plot
- Correlation Matrix (Heatmap)
- Pair Plot

```python
# Scatter Plot ‚Äî compares two continuous variables - Dataset restaurant tips
plt.figure(figsize=(8, 5))
sns.scatterplot(x='total_bill', y='tip', data=df, hue='time', style='smoker', s=100)
plt.title('üå∏ Scatter Plot ‚Äî Total Bill vs Tip\nby Time and Smoker Status')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()

print("\n‚úÖ Scatter plot displayed successfully!")
```

```python
# Correlation Heatmap ‚Äî shows relationships among multiple variables - use the dataset restaurant tips
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("üî• Correlation Heatmap ‚Äî Restaurant Tips Features")
plt.show()

print("\n‚úÖ Correlation heatmap displayed successfully!")
```

```python

# Pair Plot ‚Äî shows relationships between all numeric columns - use the dataset restaurant tips
sns.pairplot(df, hue="time")
plt.suptitle("ü™∑ Pair Plot ‚Äî Restaurant Tips Dataset", y=1.02)
plt.show()
```
