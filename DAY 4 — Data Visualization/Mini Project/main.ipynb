#### üß† MINI PROJECT ‚Äî IRIS DATA VISUALIZATION
üéØ Goal:  
Compare Sepal and Petal Lengths across species in the    
Restaurant Tips dataset using beginner-friendly visualizations.


```python
# Step 0: Import Libraries for Data Visualization and Analysis (Seaborn, Matplotlib, Pandas)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("‚úÖ Libraries imported successfully!")
```

#### 1Ô∏è‚É£ LOAD DATASET

```python
# Step 1.0: Load the Restaurant Tips dataset from CSV
csv_path = "../Dataset/Restaurant-Tips.csv"
df = pd.read_csv(csv_path)

print("\n‚úÖ Dataset loaded successfully!")
print(df.head())
```

### 2Ô∏è‚É£ UNIVARIATE ANALYSIS ‚Äî Individual Features
Analyze each feature separately to understand distribution

```python
# Step 2.0: Visualize Total Bill Distribution
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='total_bill', bins=15, kde=True, color='lightgreen')
plt.title("üå∏ Histogram ‚Äî Total Bill Distribution")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()

print("\n‚úÖ Histogram displayed successfully!")
```

```python
# Step 2.1: Density Plot for Total Bill - Restaurant Tips dataset
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='total_bill', fill=True, color='lightgreen', alpha=0.7)
plt.title("üå∏ Density Plot ‚Äî Total Bill Distribution")
plt.xlabel("Total Bill")
plt.show()

print("\n‚úÖ Density plot displayed successfully!")
```

```python

# Sepal Width Boxplot - Restaurant Tips dataset
plt.figure(figsize=(8, 5))
sns.boxplot(y=df["total_bill"] , color='lightgreen')
plt.title("üì¶ Box Plot ‚Äî Total Bill"
    "\n (Using Kernel Density Estimation)")
plt.xlabel("Total Bill")
plt.show()

print("\n‚úÖ Box plot displayed successfully!")
```

### 3Ô∏è‚É£ MULTIVARIATE ANALYSIS ‚Äî Compare Features by Species

```python
#Step 3.0: Sepal Length by Species (Boxplot) - Restaurant Tips dataset
plt.figure(figsize=(8, 5))
sns.boxplot(x='day', y='total_bill', data=df, palette='pastel')
plt.title("üìè Total Bill by Day"
          "\n (Using Boxplot Visualization)"
          "\n (Using Kernel Density Estimation)")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()

print("\n‚úÖ Box plot displayed successfully!")
```

```python
# Step 3.1: Petal Length by Species (Boxplot) - Restaurant Tips dataset
plt.figure(figsize=(8, 5))
sns.boxplot(x='time', y='total_bill', data=df, palette='pastel')
plt.title("üìè Total Bill by Time of Day"
          "\n (Using Boxplot Visualization)"
          "\n (Using Kernel Density Estimation)")
plt.xlabel("Time of Day")
plt.ylabel("Total Bill")
plt.show()

print("\n‚úÖ Box plot displayed successfully!")
```

```python
# Step 3.2: Sepal vs Petal Length Scatterplot colored by Species - Restaurant Tips dataset
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', hue='day', data=df, s=100)
plt.title("üå∏ Total Bill vs Tip by Day")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.legend(title="Day")
plt.show()

print("\n‚úÖ Scatterplot displayed successfully!")
```

#### 4Ô∏è‚É£ MULTIVARIATE CORRELATION ‚Äî Feature Relationships


```python
# step 4: Correlation Heatmap for restaurant tips dataset
plt.figure(figsize=(8, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("üî• Correlation Heatmap ‚Äî Restaurant Tips Features")
plt.show()

print("\n‚úÖ Correlation heatmap displayed successfully!")
```

#### 5Ô∏è‚É£ PAIR PLOT ‚Äî All Numeric Features by Species


```python
# Step 5.0: Pair Plot for Restaurant Tips dataset
sns.pairplot(df)
plt.suptitle("ü™∑ Pair Plot ‚Äî Restaurant Tips Dataset", y=1.02)
plt.show()

print("\n‚úÖ Pair plot displayed successfully!")
```
