```python
# ğŸ§  Mini Project â€” Analyze Restaurant Tips Dataset
# ğŸ¯ Goal: Find which day has the highest average tip using the pandas library.

# ğŸªœ Step-by-Step Solution
# Step 1: Import pandas

# We start by importing pandas, which helps us handle and analyze data easily.
import pandas as pd

print("\nâœ… pandas imported successfully!")
```

```python
# Step 2: Load the dataset - Weâ€™ll use the â€œRestaurant Tipsâ€ dataset from the dataset folder in this project.

csv_path = "../Dataset/Restaurant-Tips.csv"
print(f"\nğŸ“‚ Loading dataset from: {csv_path}")

df = pd.read_csv(csv_path)
print("\nâœ… Dataset loaded successfully!")
```

```python
# Step 2.1 : Show the first few rows of the dataset to understand its structure.
print("\nğŸ½ï¸  First 5 Rows of the Dataset")
print(df.head())

print("\nâœ… Dataset previewed successfully!")
```

```python
# Step 3: Explore the Data

# Check general info and statistics about the dataset.
print("\nğŸ“Š Dataset Information")
print(df.info())

print("\nğŸ“ˆ Dataset Statistics")
print(df.describe())
print("\nâœ… Data exploration complete!")
```

```python
# Step 4: Check Missing Values - Make sure there are no empty (missing) cells.
print("\nğŸ” Missing Values Check")

print(df.isnull().sum())
print("\nâœ… Missing values check complete!")
```

```python

# Step 4.1 If you find missing values, you can remove them using df.dropna()
print("\nğŸ§¹ Removing Missing Values")

# Remove rows with missing values
df_clean = df.dropna()

print("Original DataFrame shape:", df.shape)
print("Cleaned DataFrame shape:", df_clean.shape)

print("\nâœ… Missing values removed successfully!")
```

```python
# Step 5: Group Data by Day - Now, letâ€™s group the data by day and calculate the average tip for each.
print("\nğŸ“… Grouping Data by Day to Calculate Average Tip")

avg_tips = df.groupby("day")["tip"].mean()

print("\nğŸ’° Average Tip per Day:")
print(avg_tips)

print("\nâœ… Data grouped and average tip calculated successfully!")
```

```python
# Step 6: Find the Best Day - Finally, letâ€™s find which day has the highest average tip.
print("\nğŸ† Finding the Day with the Highest Average Tip")

best_day = avg_tips.idxmax()  # gets the day name
best_value = avg_tips.max()   # gets the tip amount

print(f"\nThe day with the highest average tips is {best_day} (${best_value:.2f})")
```

```python
# summary - This section summarizes the analysis results for clarity.
print("\nğŸ“‹ Summary of Analysis:")
print(f"- Day with Highest Average Tip: {best_day} (${best_value:.2f})")
print("\nâœ… Analysis complete!")
```
